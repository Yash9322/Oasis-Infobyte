```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## **Q1. What is average estimated unemployment rate state wise?**

```python
Unemployment_df
```

```python
group_by_region_df= Unemployment_df.groupby(by='Region').mean().sort_values(by=' Estimated Unemployment Rate (%)',ascending=False)
group_by_region_df
```

## **Q2. How to get overall unemployment rate of specific given state?**

```python
unemp_state_df= Unemployment_df[Unemployment_df.Region=='Maharashtra']
unemp_state_df
```

```python
unemp_state_df[' Estimated Unemployment Rate (%)'].mean()
```

## **Q3.Which state has highest unemployment rate overall?**

### **Chart 1. Using Line Chart to see increase in rise of unemployment**

```python

maharashtra_df
```

```python
years=list(set(list(Unemployment_df[' Date'].dt.year)))
gbyyear_df=Unemployment_df.groupby(by='year')
mean_Unemployment_rate_overall =list(gbyyear_df[' Estimated Unemployment Rate (%)'].mean())
```

```python
mean_Unemployment_rate_overall
```

```python
sns.set_style("darkgrid")
plt.plot(years, mean_Unemployment_rate_overall, 's-b')
plt.xlabel('Year')
plt.ylabel('mean_Unemployment_rate_overall(%)')
plt.title("Unemployment rate by Year")
```
